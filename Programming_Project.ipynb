```python
# Function to detect outliers using IQR method
def detect_outliers(df):
    # Calculate the first and third quartiles (Q1 and Q3)
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    outliers = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
    
    return outliers

outliers = household_data[detect_outliers(household_data[["Kidhome"]])]
outliers
```

```python
combine_df = combine_df.loc[:, ~combine_df.columns.duplicated(keep='first')]
combine_df.columns
```

```python
# Check for missing values in the combined dataset
missing_values = pd.DataFrame()
missing_values['count'] = combine_df.isnull().sum()
missing_values['avg'] = combine_df.isnull().mean()
# Display the count of missing values for each column
print("Missing Values Count:")
missing_values
```

```python
# Check if columns with null values are numeric and calculate null proportions
numeric_columns_with_null = combine_df.columns[combine_df.isnull().any()].tolist()

for column in numeric_columns_with_null:
    if pd.api.types.is_numeric_dtype(combine_df[column]):
        null_proportion = combine_df[column].isnull().mean()
        print(f"Column: {column}, Numeric: Yes, Null Proportion: {null_proportion:.2%}")
    else:
        print(f"Column: {column}, Numeric: No")

# Optionally, you can display the count of missing values for each column
print("\nMissing Values Count:")
print(combine_df[numeric_columns_with_null].isnull().sum())
```

```python
# Replace null values with the mean for numeric columns
numeric_columns_with_null = combine_df.columns[combine_df.isnull().any()].tolist()

for column in numeric_columns_with_null:
    if pd.api.types.is_numeric_dtype(combine_df[column]):
        mean_value = combine_df[column].mean()
        combine_df[column].fillna(mean_value, inplace=True)

# Verify that null values have been filled with the mean
print("Null Values Count After Filling with Mean:")
print(combine_df[numeric_columns_with_null].isnull().sum())
```
